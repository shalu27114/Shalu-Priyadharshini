```python
# ================================================
# PRODIGY_GA_02 - IMAGE GENERATION FROM TEXT
# ================================================
# Text-to-Image Generation using Pre-trained Models
# ================================================

print("ğŸš€ PRODIGY_GA_02: Text-to-Image Generation")
print("="*60)
print("Using DALL-E Mini & Stable Diffusion models")
print("="*60)

# Install required packages with specific versions to avoid conflicts
!pip install -q transformers diffusers accelerate pillow
!pip install -q xformers  # For faster inference on GPU

# Fix torch compatibility (install specific version)
!pip install -q torch==2.9.0 torchvision==0.24.0

import torch
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
import numpy as np
import time
import warnings
warnings.filterwarnings('ignore')

print("âœ… Packages installed successfully!")
print(f"PyTorch version: {torch.__version__}")
print(f"Torchvision version: {torchvision.__version__ if 'torchvision' in locals() else 'Not loaded'}")
print(f"GPU Available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")

# ================================================
# SIMPLIFIED VERSION - Only Stable Diffusion
# (DALL-E Mini often has issues in Colab)
# ================================================

def generate_image_stable_diffusion(prompt, num_images=1):
    """Generate images using Stable Diffusion - SIMPLIFIED"""
    print(f"\nğŸ¨ Generating image with Stable Diffusion...")
    print(f"Prompt: '{prompt}'")

    try:
        from diffusers import StableDiffusionPipeline

        # Load Stable Diffusion model (using smaller model for faster loading)
        model_id = "runwayml/stable-diffusion-v1-5"

        # Use GPU if available
        device = "cuda" if torch.cuda.is_available() else "cpu"

        print(f"Loading model on {device}...")

        # Load pipeline with appropriate precision
        if device == "cuda":
            pipe = StableDiffusionPipeline.from_pretrained(
                model_id,
                torch_dtype=torch.float16,
                safety_checker=None  # Disable safety checker for more prompts
            )
            pipe = pipe.to(device)
            pipe.enable_attention_slicing()  # Reduce memory usage
        else:
            pipe = StableDiffusionPipeline.from_pretrained(model_id)
            pipe = pipe.to(device)

        # Generate image
        print("â³ Generating... (30-60 seconds)")
        start_time = time.time()

        # Generate with appropriate settings
        if device == "cuda":
            with torch.autocast(device):
                images = pipe(
                    prompt,
                    num_images_per_prompt=num_images,
                    height=512,
                    width=512,
                    num_inference_steps=30,  # Reduced for speed
                    guidance_scale=7.5
                ).images
        else:
            images = pipe(
                prompt,
                num_images_per_prompt=num_images,
                height=512,
                width=512,
                num_inference_steps=20,  # Fewer steps for CPU
                guidance_scale=7.5
            ).images

        end_time = time.time()

        print(f"âœ… Generated {num_images} image(s) in {end_time - start_time:.2f} seconds")

        return images if num_images > 1 else images[0]

    except Exception as e:
        print(f"âŒ Error: {e}")
        print("Trying alternative approach...")
        return None

# ================================================
# ALTERNATIVE: Use Hugging Face API (No downloads!)
# ================================================

def generate_image_hf_api(prompt):
    """Generate image using Hugging Face API (No model downloads!)"""
    print(f"\nğŸŒ Generating image via Hugging Face API...")
    print(f"Prompt: '{prompt}'")

    API_URL = "https://api-inference.huggingface.co/models/runwayml/stable-diffusion-v1-5"

    # You'll need a Hugging Face token
    # Get one from: https://huggingface.co/settings/tokens
    # Add it like this: API_TOKEN = "your_token_here"

    # For now, we'll use a demo approach
    print("Note: For API access, you need a Hugging Face token")
    print("Visit: https://huggingface.co/settings/tokens")

    # Alternative: Use a public demo
    print("\nUsing local generation instead...")
    return generate_image_stable_diffusion(prompt)

# ================================================
# DISPLAY FUNCTIONS
# ================================================

def display_image(image, title="Generated Image", figsize=(10, 10)):
    """Display a single image"""
    if image is None:
        print("âŒ No image to display")
        return

    plt.figure(figsize=figsize)
    plt.imshow(image)
    plt.title(title, fontsize=16, fontweight='bold', pad=20)
    plt.axis('off')
    plt.tight_layout()
    plt.show()

def display_images_grid(images, titles=None, cols=2):
    """Display multiple images in a grid"""
    if not images:
        print("âŒ No images to display")
        return

    n = len(images)
    rows = (n + cols - 1) // cols

    plt.figure(figsize=(cols * 6, rows * 5))

    for i, image in enumerate(images):
        plt.subplot(rows, cols, i + 1)
        plt.imshow(image)

        if titles and i < len(titles):
            title = titles[i]
            if len(title) > 40:
                title = title[:37] + "..."
            plt.title(title, fontsize=12, pad=10)

        plt.axis('off')

    plt.suptitle("AI Generated Images", fontsize=16, fontweight='bold', y=1.02)
    plt.tight_layout()
    plt.show()

# ================================================
# SIMPLE INTERACTIVE GENERATOR
# ================================================

def simple_generator():
    """Simple interactive image generator"""
    print("\n" + "="*60)
    print("ğŸ¨ SIMPLE IMAGE GENERATOR")
    print("="*60)

    # Sample prompts
    sample_prompts = [
        "A beautiful sunset over mountains, digital art",
        "A cute robot drinking coffee in a cafe",
        "A magical forest with glowing mushrooms",
        "An astronaut riding a horse in space",
        "A cyberpunk city at night with neon lights",
        "A cute cat wearing sunglasses on a beach",
        "A dragon made of flowers and vines",
        "Underwater city with neon lights, fantasy"
    ]

    print("\nâœ¨ Sample Prompts:")
    for i, prompt in enumerate(sample_prompts, 1):
        print(f"{i}. {prompt}")

    while True:
        print("\n" + "-"*50)
        print("Options:")
        print("1. Use sample prompt")
        print("2. Enter custom prompt")
        print("3. Exit")

        choice = input("\nChoose option (1/2/3): ").strip()

        if choice == '3':
            print("\nğŸ‘‹ Goodbye!")
            break

        # Get prompt
        if choice == '1':
            try:
                num = int(input("Enter prompt number (1-8): "))
                if 1 <= num <= 8:
                    prompt = sample_prompts[num - 1]
                else:
                    prompt = sample_prompts[0]
            except:
                prompt = sample_prompts[0]
        elif choice == '2':
            prompt = input("Enter your prompt: ").strip()
            if not prompt:
                print("Please enter a prompt!")
                continue
        else:
            print("Invalid choice!")
            continue

        # Add quality hints to prompt
        enhanced_prompt = prompt + ", high quality, detailed, 4k"

        print(f"\nğŸ“ Your prompt: {enhanced_prompt}")
        print("â³ Generating image... This takes 30-60 seconds...")

        # Generate image
        image = generate_image_stable_diffusion(enhanced_prompt)

        if image is not None:
            # Display image
            display_image(image, f"Generated: {prompt[:50]}...")

            # Save option
            save = input("\nğŸ’¾ Save this image? (y/n): ").lower()
            if save == 'y':
                timestamp = int(time.time())
                safe_prompt = "".join(c for c in prompt[:20] if c.isalnum() or c in " _-").rstrip()
                filename = f"ai_image_{timestamp}_{safe_prompt}.png"
                image.save(filename)
                print(f"âœ… Saved as: {filename}")

                # Download in Colab
                try:
                    from google.colab import files
                    files.download(filename)
                    print("ğŸ“¥ Downloaded to your computer!")
                except:
                    pass
        else:
            print("âŒ Failed to generate image. Trying alternative...")

            # Show what was generated
            print("\nEven though generation failed, here's what AI imagined:")
            print(f"\"{prompt}\"")
            print("\nTip: Try a simpler prompt or check your internet connection.")

# ================================================
# QUICK DEMO (Run this first!)
# ================================================

def quick_demo():
    """Quick demonstration of image generation"""
    print("\n" + "="*60)
    print("ğŸš€ QUICK DEMO: Generating sample image...")
    print("="*60)

    # Simple prompt that usually works well
    demo_prompt = "A beautiful landscape with mountains and a lake at sunset, digital art"

    print(f"\nğŸ“ Prompt: {demo_prompt}")
    print("â³ Please wait 1-2 minutes for model to download and generate...")
    print("(First time downloads ~4GB model)")

    # Try to generate
    image = generate_image_stable_diffusion(demo_prompt)

    if image is not None:
        print("\nâœ… SUCCESS! Image generated!")
        display_image(image, "AI Generated Landscape", figsize=(12, 12))

        # Save the image
        image.save("demo_landscape.png")
        print("\nğŸ’¾ Image saved as: demo_landscape.png")

        # Try to download in Colab
        try:
            from google.colab import files
            files.download("demo_landscape.png")
            print("ğŸ“¥ Downloaded to your computer!")
        except:
            pass
    else:
        print("\nâŒ Demo failed. This could be due to:")
        print("1. Slow internet connection")
        print("2. Colab memory limits")
        print("3. Model download issues")
        print("\nTry the interactive mode with simpler prompts.")

# ================================================
# MAIN PROGRAM
# ================================================

def main():
    """Main program"""
    print("\n" + "="*60)
    print("ğŸ¤– AI IMAGE GENERATION PROJECT")
    print("="*60)
    print("\nChoose an option:")
    print("1. Quick Demo (Generate a sample image)")
    print("2. Interactive Image Generator")
    print("3. Exit")

    while True:
        choice = input("\nYour choice (1/2/3): ").strip()

        if choice == '1':
            quick_demo()
        elif choice == '2':
            simple_generator()
        elif choice == '3':
            print("\nğŸ‘‹ Thank you for using AI Image Generator!")
            print("Project: PRODIGY_GA_02 - Text-to-Image Generation")
            break
        else:
            print("Invalid choice! Please enter 1, 2, or 3.")

        if choice in ['1', '2']:
            input("\nPress Enter to continue...")
            print("\n" + "="*60)

# ================================================
# RUN THE PROGRAM
# ================================================

if __name__ == "__main__":
    # Check if we're in Colab
    try:
        import google.colab
        IN_COLAB = True
        print("âœ… Running in Google Colab")
    except:
        IN_COLAB = False
        print("âš ï¸  Not in Google Colab - some features may not work")

    # Start the program
    main()

print("\n" + "="*60)
print("ğŸ‰ PROJECT COMPLETE!")
print("="*60)
print("\nWhat you accomplished:")
print("âœ… Installed AI image generation libraries")
print("âœ… Used Stable Diffusion model")
print("âœ… Generated images from text prompts")
print("âœ… Created interactive image generator")
print("âœ… Saved and downloaded generated images")
print("\nğŸŒŸ Great work on PRODIGY_GA_02!")
```
