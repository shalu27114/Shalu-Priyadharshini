```python
# ============================================
# PRODIGY_GA_02 - Image Generation Project
# Simple & Clean Version
# ============================================

print("🚀 Starting Image Generation Project...")

# Install packages
!pip install -q torch torchvision
!pip install -q transformers diffusers accelerate

import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
from PIL import Image
import time

print(f"✅ PyTorch: {torch.__version__}")
print(f"✅ GPU: {torch.cuda.is_available()}")

# Simple image generation function
def generate_image(prompt):
    """Generate image from text prompt"""
    print(f"\n🎨 Generating: '{prompt}'")
    print("⏳ Loading model... (first time: 2-3 minutes)")

    # Load Stable Diffusion
    model = StableDiffusionPipeline.from_pretrained(
        "runwayml/stable-diffusion-v1-5",
        torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32
    )

    # Move to GPU if available
    if torch.cuda.is_available():
        model = model.to("cuda")
        model.enable_attention_slicing()

    print("✅ Model loaded! Generating image...")

    # Generate
    start = time.time()
    image = model(prompt, num_inference_steps=30).images[0]
    end = time.time()

    print(f"✅ Generated in {end-start:.1f} seconds!")

    # Display
    plt.figure(figsize=(10, 10))
    plt.imshow(image)
    plt.title(prompt, fontsize=14, pad=20)
    plt.axis('off')
    plt.show()

    # Save
    filename = f"image_{int(time.time())}.png"
    image.save(filename)
    print(f"💾 Saved as: {filename}")

    return image

# ============================================
# INTERACTIVE TEST
# ============================================

print("\n" + "="*50)
print("🤖 AI IMAGE GENERATOR")
print("="*50)

# Sample prompts
prompts = [
    "A beautiful sunset over mountains, digital art",
    "A cute robot drinking coffee in a cafe",
    "A magical forest with glowing mushrooms",
    "An astronaut riding a horse in space"
]

print("\n✨ Try these prompts or enter your own:")
for i, p in enumerate(prompts, 1):
    print(f"{i}. {p}")

while True:
    print("\n" + "-"*40)
    choice = input("\nChoose (1-4) or type 'quit': ").strip()

    if choice.lower() == 'quit':
        print("\n👋 Goodbye!")
        break

    if choice in ['1', '2', '3', '4']:
        prompt = prompts[int(choice)-1]
    else:
        prompt = input("Enter your prompt: ").strip()
        if not prompt:
            print("Please enter a prompt!")
            continue

    # Add quality hints
    enhanced_prompt = prompt + ", high quality, detailed, 4k"

    try:
        generate_image(enhanced_prompt)
    except Exception as e:
        print(f"❌ Error: {e}")
        print("Tip: Try a simpler prompt!")

    print("\nReady for next prompt!")

print("\n" + "="*50)
print("✅ PROJECT COMPLETE!")
print("="*50)
```
